# Step 2. Importing the Data Set

```python
from pandas import read_csv
filename = "BBCN.csv"
dataframe = read_csv(filename)
```

# Step 3. Split out the output variable 

The first thing we do is create an array

array= has the data set of same type.

Diff bw array and lists- array has data set of same kind while list has data of different kinds

```python
array = dataframe.values
```

The code below is the trickiest part of excercise. Now you are assigning X and Y as output variabes.

Array starts with index 0

If you take a look at dataframe you will find that it has 12 rows

On X array we are saying .. include all items in array from 0 to 11

On Y array we are saying .. just use column in array mapped to 11th column The bike buyer column.

Before we split X and Y we are going to put them in array

```python
X = array[:, 0:11]
Y = array[:, 11]
```

# Step 4. Build the Model

We can piece it all together by adding each layer

The  first layer has 11 neurons and expects 11 inputs

The second hidden layer has 8 neurons 

The third hidden layer has 8 neurons

The output layer has 1 layer to predict the output

How many hidden layers in the model?


```python
model = Sequential()
model.add(Dense(11, input_dim=11, init = "uniform", activation = "relu"))
model.add(Dense(8, init = "uniform", activation = "relu"))
model.add(Dense(8, init = "uniform", activation = "relu"))
model.add(Dense(8, init = "uniform", activation = "relu"))
model.add(Dense(8, init = "uniform", activation = "relu"))
model.add(Dense(1, init = "uniform", activation = "sigmoid"))
```

# Step 5. Compile our Model

A metric s a function that is used to judge performance of our model. Metric functions are to be supplied in metrics parameter when a model is compiled.

Lastly we set the cost or loss function to categorical_crossentropy. The binary cross_entropyis just the technical ter for the cost function in logistic regression. And categorical cross entropy is its generaliztion in for multiclass prediction via softmax


```python
model.compile(loss = "binary_crossentropy", optimizer = "adam", metrics = ["accuracy"])
```

# Step 6. Fitting data model

```python
model.fit(X, Y, nb_epoch= 1000, batch_size = 10)
```

# Step 5. Score the model

```python
score = model.evaluate(X, Y)
print("%s :%.2f%%" % (model.metrics_names[1], score[1]*100))
```
